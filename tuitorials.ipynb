**Filering Data & Regular Expressions**\
Regular Expressions can very complex and powerful for filtering/searcing data

```python
bios.loc[bios['height_cm'] > 215, ["name", "height_cm", "weight_kg"]]
# Another way of doing the same thing
bios[bios['height_cm'] > 215][["name", "height_cm", "weight_kg"]]

# Applying multiple conditions
bios[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')]
```

```python
# str.contains()
bios[bios['name'].str.contains("Keith")] # Case sensitive by default
bios[bios['name'].str.contains("keith", case = False)] # Removed case sensitivity
bios[bios['name'].str.contains("keith|Patrick", case = False)] # Multiple conditions in cotains
```

```python
# .isin()
bios[bios['born_country'].isin(["USA", "FRA", "GBR"])]
```

```python
# query function
bios.query('born_country == "USA" and born_city == "Seattle"')
```

### Adding / Removing Columns

```python
# Adding new column(s)
coffshop['Price'] = 4.99 
coffshop
```

```python
# Adding a new column with different different data for each index.
import numpy as np
coffshop['new_price'] = np.where(coffshop['Coffee Type'] == 'Espresso', 3.99, 5.99)
coffshop
```

```python
# Dropping (i.e. removing) row or column
coffshop.drop(0) # dropping a index i.e. row. Remember, it won't change the original data frame
coffshop.drop(columns='Price') # dropping a column. Rememner, ^
coffshopCopy = coffshop.copy()
coffshopCopy.drop(columns='Price', inplace=True) # Now it will drop the column permanently.
coffshopCopy
```

```python
coffshopCopy['revenue'] = coffshopCopy['Units Sold'] * coffshopCopy['new_price']
coffshopCopy
```

```python
# Renaming row/column
coffshopCopy.rename(columns={'new_price': 'price'}) # won't change the original data frame unless we set inplace = True
coffshopCopy
```

```python
coffshopCopy.rename(columns={'new_price': 'price'}, inplace=True) # will change the original data frame
# Alternative way
# coffshopCopy = coffshopCopy.rename(columns={'new_price': 'price'})
coffshopCopy
```

```python
bios_cp = bios.copy()
bios_cp['first_name'] = bios_cp['name'].str.split(' ').str[0]
bios_cp
```

```python
# Converting string date/time to date-time data type
bios_cp['born_date(dtime)'] = pd.to_datetime(bios_cp['born_date'])
bios.info()
```

```python
# Take the year from the date-year object
bios_cp['born_year'] = bios_cp['born_date(dtime)'].dt.year
bios_cp['born_year'] = bios_cp['born_year'].convert_dtypes(int)
bios_cp
```

```python
# Saving the newly created dataFrame as CSV file
bios_cp.to_csv('./complete-pandas-tutorial/data/bios_new.csv', index=False) # index=False means, we don't need index col.
```

Applying Custom Functions on the data of the DataFrame

```python
# Applying custom functions on the dataFrame using apply funcion
bios_cp['height_category'] = bios_cp['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))

# lambda function: The lambda function in Python Pandas is a small anonymous function primarily used for quick data 
# manipulation within the Pandas library. These anonymous or inline functions offer a compact way to perform operations 
# on DataFrames and Series without any formal function declaration.

bios_cp
```

```python
def categorized_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavywight'


bios_cp['Category'] = bios_cp.apply(categorized_athlete, axis=1)
bios_cp
```

### Handling NULL values

```python
coffshop.loc[[1, 2], 'Units Sold'] = np.nan
```

```python
coffshop.isna().sum() # just to get an overall idea about which columns have Null values and how many.
```

```python
# fillna function

coffshop = coffshop.fillna(coffshop['Units Sold'].mean()) # filling up NaN of "Units Sold" with the mean of that col.
coffshop
```

```python
# interpolate() function

# coffshop.loc[[2, 3], 'Units Sold'] = np.nan  # To artifically make the NaN values
new_units_sold = coffshop["Units Sold"].interpolate() 
# this function tries to continue the flow of pattern of the data 
# and changes NaN values accordingly to keep the flow maintained and returns a new column of of data
coffshop["Units Sold"] = new_units_sold
coffshop
```

```python
# Drop the NaN rows
coffshop.loc[[2, 3], 'Units Sold'] = np.nan  # To artifically make the NaN values

coffshop.dropna()
```

```python
# two other useful mathods
# coffshop.loc[[2, 3], 'Units Sold'] = np.nan  # To artifically make the NaN values
coffshop[coffshop["Units Sold"].isna()] # will give the rows that have NaN values in the specified column
coffshop[coffshop["Units Sold"].notna()] # will give the rows that DON'T have NaN values in the specified column
```
